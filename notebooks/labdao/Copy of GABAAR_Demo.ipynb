```python
import requests
import os

def download_file(url, directory, filename=None):
    local_filename = filename if filename else url.split('/')[-1]
    with requests.get(url, stream=True) as r:
        r.raise_for_status()
        with open(os.path.join(directory, local_filename), 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):
                f.write(chunk)
    return local_filename

url = 'https://github.com/labdao/plex/blob/main/testdata/sequencing/fastqc/testdata.fastq' # demo FASTQ file
fastq_filepath = download_file(url, PROJECT_DIR)

tool_url = "https://github.com/labdao/plex/blob/main/tools/fastqc/fastqc.json"
tool_path = download_file(tool_url, PROJECT_DIR)
```
