```python
#@title GO Functional Analysis of Clustering Results

pip install biopython pandas

import pandas as pd
from Bio import ExPASy
from Bio import SwissProt

# List of proteins for each cluster
proteins_cluster0 = [...]  # Replace with list of proteins for cluster 0
proteins_cluster1 = [...]  # Replace with list of proteins for cluster 1
proteins_cluster2 = [...]  # Replace with list of proteins for cluster 2
proteins_cluster3 = [...]  # Replace with list of proteins for cluster 3

clusters = [proteins_cluster0, proteins_cluster1, proteins_cluster2, proteins_cluster3]

# Fetch GO terms for each protein in each cluster
go_terms_clusters = []
for i, cluster in enumerate(clusters):
    go_terms_cluster = []
    for protein in cluster:
        handle = ExPASy.get_sprot_raw(protein)
        record = SwissProt.read(handle)
        go_terms_protein = [cross_ref[2].split(':')[1] for cross_ref in record.cross_references if cross_ref[0] == 'GO']
        go_terms_cluster.append(go_terms_protein)
    go_terms_clusters.append(go_terms_cluster)

# Convert to DataFrame
df = pd.DataFrame(go_terms_clusters, index=['Cluster 0', 'Cluster 1', 'Cluster 2', 'Cluster 3']).T
print(df)
```
